```python
df=df.drop(['model'],axis=1)
df1=df.transform(lambda x: x**0.5)
x=df1.drop(['mpg'],axis=1)
y=df1.mpg
```

```python
df1.head()
```

```python
# By applying square root transform the model gets more better
x_constant=sm.add_constant(x)
x_constant.head()

lin_reg=sm.OLS(y,x_constant).fit()
```

```python
lin_reg.summary()
```

```python
import seaborn as sns
sns.distplot(lin_reg.resid)
```

```python
import numpy as np
df2=df.transform(lambda x:np.cbrt(x))
x=df2.drop(['mpg'],axis=1)
y=df2.mpg
```

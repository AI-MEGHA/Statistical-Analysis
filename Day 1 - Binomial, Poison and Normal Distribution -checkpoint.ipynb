```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

# None of the LED are defective --- Example of Binomial Dist -- Only 2 Outcomes.

```python
p = 0.05 # Failure rate by history is 5% 
n = 10 # Sample Size
k = np.arange(0,11) # Plotting for x=0, x=1, x=2 ...
```

```python
binomial = stats.binom.pmf(k,n,p)
binomial
```

```python
plt.plot(binomial)
```

```python
print("None of the bulbs are defective is %1.4f"%binomial[0])
```

# Two or fewer defective

```python
print("Two or below 2 are defective is %1.4f"%(binomial[0]+binomial[1]+binomial[2]))
```

# The other way is to find Cumulative Probability Dist

```python
cumbinomial = stats.binom.cdf(k,n,p)
cumbinomial
```

```python
print("Two or below 2 are defective is %1.4f"%cumbinomial[2])
```

# Probability 3 and above Defective

```python
print("Three and above are defective is %1.4f"%(1-cumbinomial[2]))
```

```python
plt.plot(k,binomial,'*--')
plt.title('Binomial Dist')
plt.xlabel("No. of LED's Defective")
plt.ylabel("Probability")
```

```python
plt.stem(k,binomial,'*--')
plt.title('Binomial Dist')
plt.xlabel("No. of LED's Defective")
plt.ylabel("Probability")
```

# Normal Distribution

Steps:
    1. Convert to Z-Scale.
    2. By default the NORM function gives the area on the left side.

```python
mu = 68 # minutes
sigma = 12
```

```python
z = (90-mu)/sigma # X = 90
p = 1 - stats.norm.cdf(z)
print('The proportion of smart phone users spending more than 90 mins is %1.4f'%p)
```

```python
z = (20-mu)/sigma # X = 20
p = stats.norm.cdf(z)
print('The proportion of smart phone users spending less than 20 mins is %1.8f'%p)
```

```python
z1 = (50-mu)/sigma # X = 50
z2 = (100-mu)/sigma # X = 100
p1 = stats.norm.cdf(z1)
p2 = stats.norm.cdf(z2)
print('The proportion of smart phone users spending between 50 and 100 mins is %1.4f'%(p2-p1))
```

# Hypothesis Testing

## One Sample t test

```python
a = np.random.randn(70)
s_std = (a*0.6)+2.8 # Making use of Z scaling formula
s_std
```

```python
s_std.mean()
```

```python
s_std.std()
```

## In Built funs

```python
import scipy.stats as stats
test_stat,pval = stats.ttest_1samp(s_std,3)
(test_stat,pval)
```

### As t-test value is GREATER tha 2.58, we can clearly reject the null hypotheis ... and GIVE THE OWNER a PLATINUM CERT as he is performing at more than 99%
